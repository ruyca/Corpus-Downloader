```python
# Utilizar requirements2.txt para utilizar este codigo 
from pytubefix import YouTube
from tinydb import TinyDB, Query
import os
import time
```

```python
captions_dir = os.path.join(os.getcwd(), "captions")
```

```python
# Cargamos la base de datos
db = TinyDB("audio_database.json")
db.all()
```

```python
def create_captions(link: str) -> None:
    yt = YouTube(link)
    vid_id = yt.video_id
    try: 
        caption = yt.captions.get_by_language_code("a.es")
        file_path = os.path.join(captions_dir, f"{vid_id}.txt")
        caption.save_captions(file_path)
    except:
        print(f"No se pudo descargar los subtitulos del video: {vid_id}")
```

```python
base_link = "https://www.youtube.com/watch?v="
link = ""
timer = 0
for item in db: 
    link = base_link + item["VideoID"]
    if timer % 5 == 0:
        time.sleep(30)
    create_captions(link)
    timer += 1

print("\n\n**************DONE****************")
```
